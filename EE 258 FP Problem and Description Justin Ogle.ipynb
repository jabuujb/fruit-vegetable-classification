```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
```

```python
dataset_path = 'Desktop/Grad School Homework/EE 258/EE 258 Final Project/Final Project Data'
```

```python
# List the class directories inside the dataset folder
class_names = os.listdir(dataset_path)

# Define the number of subfolders to display per class
num_subfolders_per_class = 3

# Visualize subfolders within each class
for class_name in class_names:
    class_path = os.path.join(dataset_path, class_name)
    
    # List subfolders within the class folder
    subfolders = [subfolder for subfolder in os.listdir(class_path) if os.path.isdir(os.path.join(class_path, subfolder))]
    
    if subfolders:
        plt.figure(figsize=(15, 5))
        plt.suptitle(f"Subfolders in Class: {class_name}", fontsize=16)
        
        for i, subfolder_name in enumerate(subfolders[:num_subfolders_per_class]):
            subfolder_path = os.path.join(class_path, subfolder_name)
            sample_images = os.listdir(subfolder_path)
            
            for j, image_name in enumerate(sample_images[:3]):
                image_path = os.path.join(subfolder_path, image_name)
                image = Image.open(image_path)
                
                plt.subplot(num_subfolders_per_class, 3, i * 3 + j + 1)
                plt.imshow(image)
                plt.title(subfolder_name)
                plt.axis('off')
        
        plt.tight_layout()
        plt.show()
```

```python
#### This visualization tool output displays the first three classes and its first three images within the three Subfolders.
#### The outlier would be the train data as it does not have the same images as the test and validation. 
#### This outlier does makes sense because train data contains more and random amounts of images and is suppose to run agains the test and validation images.
```

```python
# List the class directories inside the dataset folder
class_names = os.listdir(dataset_path)

# Visualize and describe datasets in subfolders for each class
for class_name in class_names:
    class_path = os.path.join(dataset_path, class_name)
    
    # List subfolders within the class folder
    subfolders = [subfolder for subfolder in os.listdir(class_path) if os.path.isdir(os.path.join(class_path, subfolder))]
    
    if subfolders:
        print(f"Class: {class_name}")
        
        # Initialize lists to store subfolder names and corresponding image counts
        subfolder_names = []
        image_counts = []
        
        for subfolder_name in subfolders:
            subfolder_path = os.path.join(class_path, subfolder_name)
            subfolder_images = os.listdir(subfolder_path)
            num_subfolder_images = len(subfolder_images)
            
            # Append subfolder name and image count to lists
            subfolder_names.append(subfolder_name)
            image_counts.append(num_subfolder_images)
            
            print(f"Subfolder: {subfolder_name}")
            print(f"Number of Images: {num_subfolder_images}")
            
        # Create a bar plot of image counts per subfolder
        plt.figure(figsize=(10, 6))
        sns.barplot(x=subfolder_names, y=image_counts)
        plt.title(f"Image Counts in Subfolders for Class: {class_name}")
        plt.xlabel("Subfolder Name")
        plt.ylabel("Number of Images")
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()
            
        print("-" * 30)
        
        # Create a scatter plot of subfolder names vs. image counts
        plt.figure(figsize=(10, 6))
        plt.scatter(subfolder_names, image_counts, color='blue', marker='o')
        plt.title(f"Image Counts in Subfolders for Class: {class_name}")
        plt.xlabel("Subfolder Name")
        plt.ylabel("Number of Images")
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()
            
        print("-" * 30)
```
